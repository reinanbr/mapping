```python
import osmnx as ox
import matplotlib.pyplot as plt
import numpy as np
```

```python
gdf_pet = ox.geocode_to_gdf([{'city':'petrolina'},{'city':'juazeiro'},{'city':'Paulo Afonso'}])
```

```python
gdf_pet.head()
```

```python
plt.figure()
fig, ax = plt.subplots(1, figsize=(10, 6))



mp = gdf_pet.plot(column='importance',cmap='magma',linewidth=0.8, ax=ax, edgecolor='0.8')


v_max,v_min = gdf_pet['importance'].max(),gdf_pet['importance'].min()

sm = plt.cm.ScalarMappable(cmap='magma', norm=plt.Normalize(vmin=v_min, vmax=v_max))
sm._A = []

cbar = fig.colorbar(sm)

plt.title('oi')



```
